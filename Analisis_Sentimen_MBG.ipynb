##**Labeling dengan Groq (LLaMA)**

```python
!pip install groq pandas python-dotenv
```

```python
import time
from groq import Groq

client = Groq(api_key=os.environ.get("GROQ_API_KEY"))
positif_kw = [
    'bagus', 'setuju', 'dukung', 'mantap', 'lanjutkan',
    'terimakasih', 'alhamdulillah', 'semoga sukses', 'bermanfaat',
    'terima kasih pak', 'salut', 'hebat', 'luar biasa', 'bangga',
    'semangat pak', 'support', 'pro rakyat', 'kerja bagus',
    'program bagus', 'masyaallah', 'keren'
]

negatif_kw = [
    'tutup selamanya', 'stop mbg', 'hentikan', 'buang buang',
    'korupsi', 'gagal', 'tidak berguna', 'gak berguna',
    'buang anggaran', 'mending ganti', 'percuma', 'bohong',
    'tipu', 'kecewa', 'tidak merata', 'gak merata',
    'sembako mahal', 'harga mahal', 'beras mahal'
]

neutral_kw = [
    'kenapa di hentikan', 'kenapa sementara', 'bagaimana dengan',
    'kapan', 'di mana', 'gimana caranya', 'kalbar', 'lampung',
    'pertanyaan', 'info dong', 'tanya'
]

def rule_based(text):
    t = str(text).lower()
    for kw in negatif_kw:
        if kw in t:
            return 'negative'
    for kw in positif_kw:
        if kw in t:
            return 'positive'
    for kw in neutral_kw:
        if kw in t:
            return 'neutral'
    return None

SYSTEM_PROMPT = """
Kamu adalah analis sentimen komentar TikTok Indonesia tentang program MBG (Makan Bergizi Gratis).

Klasifikasikan ke salah satu:
- positive: mendukung MBG, pujian ke pemerintah, apresiasi program
- negative: menolak MBG, kritik program, sindir pemerintah, minta dihentikan
- neutral: pertanyaan, informasi, laporan kondisi daerah, tidak jelas arahnya

ATURAN PENTING:
- "tutup", "stop", "hentikan" = negative
- "lebih baik sekolah gratis" = negative (mengkritik dengan perbandingan)
- "gara-gara MBG harga mahal" = negative (menyalahkan program)
- "kenapa dihentikan sementara" = neutral (pertanyaan, belum tentu pro/kontra)
- Komentar panjang objektif menjelaskan program = neutral atau positive
- Emoji 😭 bisa berarti frustrasi, perhatikan konteks kalimat
- Jawab HANYA dengan satu kata: positive, negative, atau neutral
""".strip()

def label_groq(text):
    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user",   "content": str(text)}
            ],
            max_tokens=5,
            temperature=0
        )
        result = response.choices[0].message.content.strip().lower()
        return result if result in ['positive', 'negative', 'neutral'] else 'neutral'
    except Exception as e:
        print(f"Error: {e}")
        time.sleep(5)
        return 'neutral'

print("Memulai labeling Groq...")
labels_groq = []
groq_count  = 0

for i, row in df.iterrows():
    label = rule_based(row['komentar'])
    if label is None:
        label = label_groq(row['komentar'])
        groq_count += 1
        time.sleep(0.1)
    labels_groq.append(label)
    if (i + 1) % 50 == 0:
        print(f"  Progress: {i+1}/{len(df)} | Groq calls: {groq_count}")

df['label_groq'] = labels_groq

print(f"\n Groq selesai! API calls: {groq_count}")
print(df['label_groq'].value_counts())
```

## **labeling dengan indoBERT**

```python
!pip install transformers torch
```

```python
from transformers import pipeline

print("Loading IndoBERT... (tunggu sebentar)")

indobert = pipeline(
    "text-classification",
    model="w11wo/indonesian-roberta-base-sentiment-classifier",
    tokenizer="w11wo/indonesian-roberta-base-sentiment-classifier"
)

label_map = {
    'POSITIVE' : 'positive',
    'NEGATIVE' : 'negative',
    'NEUTRAL'  : 'neutral'
}

def label_indobert(text):
    try:
        result = indobert(str(text)[:512])[0]
        return label_map.get(result['label'], 'neutral'), round(result['score'], 4)
    except:
        return 'neutral', 0.0

print("Memulai labeling IndoBERT...")
labels_bert  = []
scores_bert  = []

for i, row in df.iterrows():
    label, score = label_indobert(row['komentar'])
    labels_bert.append(label)
    scores_bert.append(score)
    if (i + 1) % 50 == 0:
        print(f"  Progress: {i+1}/{len(df)}")

df['label_indobert']       = labels_bert
df['confidence_indobert']  = scores_bert

print(f"\n IndoBERT selesai!")
print(df['label_indobert'].value_counts())
```

## **Override IndoBERT dengan rule-based**

```python
def fix_indobert(row):
    text = str(row['komentar']).lower()
    label = row['label_indobert']

    # Kata yang JELAS negatif tapi IndoBERT salah tangkap
    negatif_override = [
        'tutup', 'stop', 'hentikan', 'buang', 'korupsi',
        'gagal', 'percuma', 'bohong', 'tipu', 'kecewa',
        'mahal', 'mending ganti', 'tidak berguna', 'gak berguna',
        'sia sia', 'ganti program', 'lebih baik sekolah gratis'
    ]

    # Kata yang JELAS positif
    positif_override = [
        'lanjutkan', 'dukung', 'setuju', 'bagus', 'bermanfaat',
        'terima kasih pak', 'terimakasih', 'salut', 'hebat',
        'semangat pak', 'pro rakyat', 'alhamdulillah'
    ]

    for kw in negatif_override:
        if kw in text:
            return 'negative'
    for kw in positif_override:
        if kw in text:
            return 'positive'

    return label  # biarkan IndoBERT kalau tidak ada override

df['label_indobert_fixed'] = df.apply(fix_indobert, axis=1)

print("Sebelum fix:")
print(df['label_indobert'].value_counts())
print("\nSetelah fix:")
print(df['label_indobert_fixed'].value_counts())
```

## **Hybrid Label**

```python
def hybrid_label(row):
        if row['label_groq'] == 'neutral' and row['label_indobert_fixed'] != 'neutral':
        if row['confidence_indobert'] >= 0.80:
            return row['label_indobert_fixed']
    return row['label_groq']

merged_df['label_final'] = merged_df.apply(hybrid_label, axis=1)

print("Distribusi label final (Hybrid):")
print(merged_df['label_final'].value_counts())
print()

berubah = (merged_df['label_final'] != merged_df['label_groq']).sum()
print(f"Komentar yang dikoreksi IndoBERT : {berubah} baris")
print(f"Tetap pakai Groq                 : {len(df) - berubah} baris")

print("\n=== Contoh yang dikoreksi IndoBERT ===")
changed = merged_df[merged_df['label_final'] != merged_df['label_groq']][
    ['komentar', 'label_groq', 'label_indobert_fixed',
     'confidence_indobert', 'label_final']
].head(15)
print(changed.to_string())

merged_df.to_csv("Sentimen_MBG_Labeled.csv", index=False)
print("\n Tersimpan: Sentimen_MBG_Labeled.csv")
```

```python
import pandas as pd
from sklearn.metrics import (classification_report,
                              confusion_matrix, accuracy_score)
import seaborn as sns
import matplotlib.pyplot as plt

eval = pd.read_csv('/content/labeling manual.csv', on_bad_lines='skip')

eval['label_manual'] = eval['label_manual'].str.strip().str.lower()
eval['label_manual'] = eval['label_manual'].replace({'neutal': 'neutral'})
eval['label_model']  = eval['label_model'].str.strip().str.lower()

eval.dropna(subset=['label_model', 'label_manual'], inplace=True)

print(f"Total data evaluasi: {len(eval)} baris")
print()
print("Distribusi label_model :")
print(eval['label_model'].value_counts())
print()
print("Distribusi label_manual:")
print(eval['label_manual'].value_counts())

y_true = eval['label_manual']
y_pred = eval['label_model']

acc = accuracy_score(y_true, y_pred)
print(f"\n{'='*45}")
print(f"  Akurasi Model Hybrid : {acc*100:.2f}%")
print(f"{'='*45}")

print("\n=== Classification Report ===")
print(classification_report(y_true, y_pred,
      target_names=['negative', 'neutral', 'positive'],
      zero_division=0))

labels = ['positive', 'negative', 'neutral']
cm = confusion_matrix(y_true, y_pred, labels=labels)

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=labels, yticklabels=labels,
            linewidths=0.5, ax=ax)
ax.set_xlabel('Prediksi Model', fontsize=12)
ax.set_ylabel('Label Manual (Ground Truth)', fontsize=12)
ax.set_title(f'Confusion Matrix — Model Hybrid\nAkurasi: {acc*100:.2f}%',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrix_evaluasi.png', dpi=150)
plt.show()
print("📊 Tersimpan: confusion_matrix_evaluasi.png")


eval['benar'] = y_true == y_pred
salah = eval[~eval['benar']][['komentar', 'label_manual', 'label_model']]
print(f"\n=== Komentar yang salah prediksi ({len(salah)} baris) ===")
print(salah.to_string())
```

Dari hasil evaluasi, hybrid label sudah baik jika digunakan pada dataset komentar MBG yang banyak mengandung teks gaul, typo, dan sarkas. Batas wajar akurasi pada dataset paling sulit adalah 60--70%.
